## 计算期刊TFIDF

```python
#定义用于tfidf的分词器
def tokench(text):
    tokens = [word for word in text.split(' ')]#数据以空格隔开
    after_tokens = []
    for token in tokens:
        if len(token)>1:
            after_tokens.append(token)
    return after_tokens
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                 tokenizer=tokench,use_idf=True)

tfidf_matrix = tfidf_vectorizer.fit_transform(jdf['title'])
```

```python
terms = tfidf_vectorizer.get_feature_names()#获取词（特征）名
```

```python
print terms[3]
```

```python
weight = tfidf_matrix.todense().tolist()#tfidf权重
```

```python
len(weight)
```

```python
tfidf = pd.DataFrame({'terms':terms, jdf['journal'][0]:weight[0],jdf['journal'][1]:weight[1],
                     jdf['journal'][2]:weight[2],jdf['journal'][3]:weight[3],
                     jdf['journal'][4]:weight[4],jdf['journal'][5]:weight[5],
                     jdf['journal'][6]:weight[6],jdf['journal'][7]:weight[7],
                     jdf['journal'][8]:weight[8],jdf['journal'][9]:weight[9],
                     jdf['journal'][10]:weight[10],jdf['journal'][11]:weight[11],
                     jdf['journal'][12]:weight[12],jdf['journal'][13]:weight[13],
                     jdf['journal'][14]:weight[14]})
```

## 构建LDA建模数据 

```python
sf = gl.SFrame(df)
```

```python
#词袋转换
bow = gl.text_analytics.count_words(sf['titlecut'])
```

```python
sf['bow'] = bow
```

```python
sf.save('jcr')
```

## LDA建模

```python
model = gl.topic_model.create(sf['bow'],num_topics=15,alpha=0.1,beta=0.1,num_iterations=100)
```

## 模型可视化

```python
from __future__ import division

import graphlab as gl
import pyLDAvis
import pyLDAvis.graphlab

pyLDAvis.enable_notebook(local=True)
```

```python
pyLDAvis.graphlab.prepare(model,sf['bow'])
```

## 预测主题概率分布

```python
#预测主题概率分布
pred = model.predict(sf['bow'], output_type='probability')
```

```python
sf['topic'] = pred
```
